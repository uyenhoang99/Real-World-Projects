```python
from datetime import datetime, timedelta
```

```python
def format_date(timestamp, datetime_format):
    try:
        time =  datetime.fromtimestamp(timestamp)
        return datetime.strftime(time, datetime_format)
    except:
        print("The argument 'timestamp' needs to be formatted as integers. The 'datetime_format' argument must present the desired date format, ex: '%d-%m-%Y'.")
```

```python
format_date(1514665153, "%d-%m-%Y")
```

```python
def calculate_landing_time(rocket_launch_dt, travel_duration):
    try:
        dt = rocket_launch_dt + timedelta(days = travel_duration)
        return datetime.strftime(dt, "%d-%m-%Y")
    except:
        print("The 'rocket_launch_dt' argument must be entered using the 'datetime()' format.")
```

```python
calculate_landing_time(datetime(2023, 2, 15), 20)
```

```python
def days_until_delivery(expected_delivery_dt, current_dt):
    try:
        return int((expected_delivery_dt - current_dt).total_seconds() / (60*60*24))
    except:
        print("The 'expected_delivery_date' argument and the 'current_dt' argument should be entered using the 'datetime()' format.")
```

```python
days_until_delivery(datetime(2023, 2, 15), datetime(2023, 2, 5))
```
