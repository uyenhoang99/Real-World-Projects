**Which occupation has the lowest average sleep duration?**

Save this in a string variable called lowest_sleep_occ.

```python
#groupby occupation for sleep duration 
sleep_duration = round(df.groupby('Occupation')['Sleep Duration'].mean(),2)
sleep_duration.to_dict()
sleep_duration.values
```

```python
#for loop to find the occupation with the lowest average sleep
lowest_sleep_occ = []
min_value = min(sleep_duration.values)

for key, value in sleep_duration.items():
    if value == min_value:
        lowest_sleep_occ.append(key)
lowest_sleep_occ
```

```python
#save lowest_sleep_occ as string
lowest_sleep_occ = lowest_sleep_occ[0]
lowest_sleep_occ
```

**Which occupation has the lowest average sleep quality?**

Save this in a string variable called lowest_sleep_quality_occ. Did the occupation with the lowest sleep duration also have the lowest sleep quality? If so assign a boolean value to variable same_occ variable, True if it is the same occupation, and False if it isn't.


```python
#average sleep quality by occupation
sleep_quality = round(df.groupby('Occupation')['Quality of Sleep'].mean(),2)
sleep_quality.to_dict()
sleep_quality.values

lowest_sleep_quality_occ = []
min_val_sleep_quality = min(sleep_quality.values)

for key, value in sleep_quality.items():
    if value == min_val_sleep_quality:
        lowest_sleep_quality_occ.append(key)
lowest_sleep_quality_occ
```

```python
#save lowest_sleep_quality_occ as string
lowest_sleep_quality_occ = lowest_sleep_quality_occ[0]
lowest_sleep_quality_occ
```

```python
#same occupation for lowest sleep duration and lowest sleep quality check
if lowest_sleep_occ == lowest_sleep_quality_occ:
    same_occ = True
else:
    same_occ = False

print(same_occ)
```
