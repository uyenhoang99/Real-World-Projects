**Which decade had the highest ratio of US-born Nobel Prize winners to total winners in all categories?**

Store this as an integer called max_decade_usa.

```python
#calculate decades
df["decade"] = (df["year"] // 10) * 10
df.head()
df["decade"].unique()
```

```python
#calculate the ratio of US-born winners versus non-US-born winners
df_us_born = df[df["birth_country"] == "United States of America"]
df_us_born = df_us_born["decade"].value_counts(sort = False)

df_non_us_born = df["decade"].value_counts(sort = False)
((df_us_born / df_non_us_born) * 100).sort_values(ascending=False)
```

```python
max_decade_usa = 2000
```

**Which decade and Nobel Prize category combination had the highest proportion of female laureates?**

Store this as a dictionary called max_female_dict where the decade is the key and the category is the value. There should only be one key:value pair.

```python
#filter for female laureates
df_f = df[df["sex"] == "Female"]
female = df_f["decade"].value_counts(sort = False)
```

```python
#calculate the ratio of female laureates to non-female laureates
all = df["decade"].value_counts(sort=False)
((female / all) * 100).sort_values(ascending = False)
```

```python
#calculate the ratio of categories won by female laureates versus non-female laureates
female_cat = df_f["category"].value_counts(sort = False)
all_cat = df["category"].value_counts(sort=False)
((female_cat/all_cat)*100).sort_values(ascending=False)
```
