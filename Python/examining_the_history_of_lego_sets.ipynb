**What percentage of all licensed sets ever released were Star Wars themed?**

Save your answer as a variable the_force, as an integer (e.g. 25).

```python
# Start coding here
# Use as many cells as you need
#find licensed themes
licensed_themes = parent_themes[parent_themes['is_licensed'] == True]
licensed_themes
```

```python
#extract all licensed themes into a list 
licensed_themes_list = list(licensed_themes['name'])
licensed_themes_list
```

```python
#find lego sets with licensed themes
sets_with_licensed_themes = lego_sets[lego_sets['parent_theme'].isin(licensed_themes_list)]
sets_with_licensed_themes
```

```python
#count the number of sets with licensed themes
sets_with_licensed_themes_count = sets_with_licensed_themes.shape[0]
sets_with_licensed_themes_count
```

```python
#count the number of licensed sets with Star Wars theme
star_wars_count = 0
for i in sets_with_licensed_themes['parent_theme']:
    if i == 'Star Wars':
        star_wars_count += 1
print(star_wars_count)
```

```python
#percentage of all licensed sets ever released were Star Wars themed
the_force = int(round((star_wars_count / sets_with_licensed_themes_count *  100), 0))
the_force
```

Percentage of all licensed sets ever released were Star Wars themed = 46%

**In which year was the highest number of Star Wars sets released?**

Save your answer as a variable new_era, as an integer (e.g. 2012).

```python
#subset for star wars themed sets
star_wars = lego_sets[lego_sets['parent_theme'] == 'Star Wars']
year = star_wars['year'].value_counts().reset_index()
year.columns = ['year','count']
year
```

```python
#year with the highest number of Star Wars sets released
new_era = int(year['year'][0])
new_era
```

The year that has the highest number of Star Wars sets released = 2016
