**What was the most frequent movie duration in the 1990s?**

Save an approximate answer as an integer called duration (use 1990 as the decade's start year).


```python
#filter for movies in the 1990s
df_90s = df[(df["release_year"] >= 1990) & (df["release_year"] < 2000)]
df_90s.head()
```

```python
#Graph the duration
#plot duration values for movies in the
plt.figure(figsize = (20,10))
sns.countplot(data = df_90s, x = "duration", order = df_90s["duration"].value_counts().index)
plt.title("Movie Duration for Movies in the 1990s")
plt.xticks(rotation = 90)
```

**A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count.**

```python
#filter for action movies
df_90s_action = df_90s[df_90s["genre"] == "Action"]
df_90s_action.head()
```

```python
#count short action movies
l = []
for index, value in df_90s_action["duration"].items():
    if value < 90:
        l.append(value)

print("There are", len(l), "action short action movies released in the 1990s.")
```
